# Time Domain

```python
%matplotlib inline
import numpy as np
import matplotlib
import matplotlib.pyplot  as plt
import math
import scipy.fftpack
```

```python
duration = eval(input("Duration (s):"))
freq =  eval(input("Frequency (Hz):"))
fs   =  eval(input("Sample Rate:"))
amp  =  eval(input("Amplitutde (V):"))
t    = np.arange(0,duration, 1/(f*fs))
s1   = amp*np.sin(2*np.pi*freq*t)
fig, ax = plt.subplots() #Fig: figure object, ax: plot axes
ax.set(xlabel='Time (s)', ylabel='Voltage (V)', title='Wavy')
ax.grid()
ax.plot(t,s1)
```
